## Dictionaries in Python

Dictionaries sind unglaublich praktisch, damit kannst du z. B. folgendes machen:

- Du kannst Wertezuordnungen speichern (z.B. Telefonbuch: Ein Nachname hat eine Telefonnummer).
- Du kannst nachträglich Elemente verändern / entfernen / hinzufügen.
- Dictionaries brauchst du wirklich immer wieder...

Machen wir mal ein Beispiel...

```python
d = {"Berlin": "BER", "Helsinki": "HEL", "Saigon": "SGN", "Budapest": "BDA"}
```

```python
print(d)
```

Zugriff auf ein einzelnes Element:

```python
print(d["Helsinki"])
```

Hiermit überschreibst du ein einzelnes Element:

```python
d["Budapest"] = "BUD"
```

```python
print(d)
```

## Dictionaries und Schleifen

Du hast bei Dictionaries 2 Möglichkeiten, diese mit einer for - Schleife durchzugehen.

Entweder direkt, dann gehst du die Schlüssel durch:

```python
d = {"München": "MUC", "Budapest": "BUD", "Helsinki": "HEL"}

for key in d: 
    value = d[key]
    print(key)
    print(value)
    print("***")
```

Oder über die .items() - Methode, damit kannst du Schlüssel + Wert direkt durchgehen:

```python
print("Key/values as tuple pair")
for key_value in swissairports.items():
    print(key_value[0] + ": " + key_value[1])
    print("***")

print("\nKey/values as single attributes")
    
for key, value in swissairports.items():
    print(key + ": " + value)
    print("***")
```

Wie oft kommt jedes Wort in der liste vor?

```python
liste = ["Hallo", "Hallo", "Welt", "Hallo", "Mars"]

from collections import Counter
count = Counter(liste)
print(count.most_common())

d = {}
for element in liste:
    if(element not in d):
        d.update({element:1})
    else:
        d.update({element:d.get(element)+1})
        
print(d)
# dein code hier 

#print(d)
```

Welche Institute hatten die meisten beiteiligten Personen?

```python
#import requests
import csv
#data = requests.get("http://p3.snf.ch/P3Export/P3_PersonExport.csv").text.splitlines() #does not work
# we open the local csv file
data = open("P3_PersonExport.egg", "r")
# we create an object that returns a list of dict's (key, value) tuple pairs for every entry in the file
personen = csv.DictReader(data,delimiter=';')

institutes = {} # dictionary with institute name and counter

#iterate through personen
for row in personen:
    
    # iterate through row
    for key, value in row.items():
        #print(key, value)
        #
        # check if we are at the right column
        if "Institute Name" == key:
            #print(key)
            # check if there is a value
            if value == None or '' == value:
                continue
            elif value not in institutes:
                #print(value)
                # if the institute is not there yet we add it 
                institutes.update({value:1})
            else:
                #print(value)
                # if the institute is there we raise the counter
                institutes.update({value:institutes.get(value)+1}) 
    #print(institute)

# we are through the iteration, now the final Schliff
# sort the results (dictonaries do not offer sorting, we have to copy paste a sorting solution from Google)
import operator
sorted_institutes = sorted(institutes.items(), key=operator.itemgetter(1))

# print the sorted dictionary
print(sorted_institutes)


```
