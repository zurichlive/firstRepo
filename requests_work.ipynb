```python
import requests
import pandas as pd
import json
import urllib.request
```

```python
basic_url = 'http://ws-old.parlament.ch/'
search = 'councillors/?'
#councillors/basicdetails (ohne paging)
fmt = 'format=json'
parameter = '&lang=de'
pagenumber = 1

print(basic_url + search + fmt + parameter)
r = requests.get(basic_url + search + fmt + parameter)
print(r.status_code)
print(r.url)
#print(r.json())

request = urllib.request.Request(basic_url + search + fmt + parameter, headers={'User-Agent': 'Mozilla'})
connection = urllib.request.urlopen(request)
js = connection.read()
info = json.loads(js.decode("utf-8"))
info_total = info
```

```python
break_operator = True

print(type(info))       #list
print(type(info[0]))    #dict
print(type(info_total)) #list

while (break_operator):
    for x in info:
        if 'hasMorePages' in x.keys():
            if(bool(x['hasMorePages'])):
                print('yes')
                request = urllib.request.Request(basic_url + search + fmt + parameter + '&pagenumber='+str(pagenumber), headers={'User-Agent': 'Mozilla'})
                connection = urllib.request.urlopen(request)
                js = connection.read()
                #info_total = info + json.loads(js.decode("utf-8"))
                info_total = info_total + json.loads(js.decode("utf-8"))
                info = json.loads(js.decode("utf-8"))
                pagenumber = pagenumber + 1
            elif not bool(x['hasMorePages']):
                print('no')
                break_operator = False
                break

data = pd.DataFrame(info_total)
data.to_csv("parlament2.csv")
```
