```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import random
import os
```

```python
url_base = 'https://www.admin.ch/gov/de/start/dokumentation/veranstaltungen.event-id-'
url_end = '.html'
```

```python
#Saving off the html code into a folder
for elem in range(6000,7000): #Onliny numbers between 6000 and 7000
    r = requests.get('https://www.admin.ch/gov/de/start/dokumentation/veranstaltungen.event-id-'+str(elem)+'.html')
    r = BeautifulSoup(r.text, 'html.parser')
    r = r.prettify("utf-8")
    with open("html/"+str(elem)+".html", "wb") as file:
        file.write(r)
        file.close()
```

```python
# Visiting the html code and making a mini table for each individual file
lst = os.listdir('html')
for htm in lst:
    f = open("html/"+htm, 'r')
    r = BeautifulSoup(f.read(), 'html.parser')
    dic = {}
    for elem in r.find_all('tr'):
        Key = elem.find('th').text.strip()
        Value = elem.find('td').text.strip()
        minidict = {Key:Value}
        dic.update(minidict)
    dic.update({'File':htm})
    pd.DataFrame([dic]).to_csv("csvs/"+htm+".csv", index=False) 
```
