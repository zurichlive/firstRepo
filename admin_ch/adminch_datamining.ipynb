```python
#Merging all of the csvs
lst = os.listdir('csvs')
df = pd.DataFrame()
for elem in lst:
    new = pd.read_csv('csvs/'+elem)
    df = pd.concat([df, new],sort=True)
df = pd.DataFrame(df)
#df
```

```python
# cleaning
del df['Unnamed: 0']
df = df.rename(columns={'Datum\n           \n           Zeit': 'Datum Zeit'})
df = df.rename(columns={'Datum\n           \n            Zeit': 'Datum Zeit2'})
```

```python
# set a proper index
df['ID'] = df['ID'].astype(int) 
df.set_index('ID', inplace=True)
df = df.sort_index()
```

```python
# Combine Datum and Datum Zeit
df['Datum'].fillna(df['Datum Zeit'], inplace=True)
df['Datum'].fillna(df['Datum Zeit2'], inplace=True)
```

```python
del df['Datum Zeit']
del df['Datum Zeit2']
```

```python
# methods for returning clean date from messy date
#
#

def dateconverter(date_raw):
    date_clean = date_raw.lstrip()
    match = re.search(r'(\d+.\d+.\d+)',date_clean)
    #print(match.group(1))
    #date_clean = date_clean[3:13]
    return match.group(1)

def yearconverter(date):
    year = date.year
    return year

def monthconverter(date):
    month = date.month
    return month

def dayconverter(date):
    day = date.day
    return day
```

```python
df['New_Date'] = df['Datum'].apply(dateconverter)
```

```python
df['New_Date'] = pd.to_datetime(df['New_Date'], format="%d.%m.%Y")
```

```python
df.dtypes
```

```python
df['Year'] = df['New_Date'].apply(yearconverter)
df['Month'] = df['New_Date'].apply(monthconverter)
df['Day'] = df['New_Date'].apply(dayconverter)
```

```python
def putznamen(veranstalter_name):
    if veranstalter_name.startswith('APK-N'):
        return 'APK-N'
    if veranstalter_name.startswith('compenswiss'):
        return 'Compenswiss'
    if veranstalter_name.startswith('Direktion f√ºr Entwicklung und Zusammenarbeit (DEZA) und Staatssekretariat'):
        return 'Direktion f√ºr Entwicklung und Zusammenarbeit (DEZA) / Staatssekretariat f√ºr Wirtschaft (SECO)'
    if veranstalter_name.startswith('Zentrum f√ºr Demokratie Aarau'):
        return 'Zentrum f√ºr Demokratie Aarau (ZDA)'
    else:
        return veranstalter_name
```

```python
df['Veranstalter'] = df['Veranstalter'].apply(putznamen)
```
