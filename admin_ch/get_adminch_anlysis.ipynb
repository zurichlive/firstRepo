```python
# return date from messy date
def dateconverter(date_raw):
    date_clean = date_raw[3:13]
    return date_clean

def yearconverter(date):
    year = date.year
    return year

def monthconverter(date):
    month = date.month
    return month

def dayconverter(date):
    day = date.day
    return day
#month = date_raw[5:6]
#day = date_raw[8:9]
```
