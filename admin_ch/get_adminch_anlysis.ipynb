```python
# return date from messy date
#
#
def dateconverter(date_raw):
    date_clean = date_raw[3:13]
    return date_clean

def yearconverter(date):
    year = date.year
    return year

def monthconverter(date):
    month = date.month
    return month

def dayconverter(date):
    day = date.day
    return day
```

```python
#df['New Date'] = pd.to_datetime(df['Date'], format="%d.%m.%Y")
#df_all['New Date'] = pd.to_datetime(df_all['Date'], format="   %d.%m.%Y")
#Mo 07.01.2008 13:30
#str(df_all['Date'].iloc[1])[3:13]
#df_all['Date'][3:12]

#del df_all['New_Date']
df_all['New_Date'] = df_all['Date'].apply(dateconverter)
df_all['New_Date'] = pd.to_datetime(df_all['New_Date'], format="%d.%m.%Y")
```

```python
df_all['Year'] = df_all['New_Date'].apply(yearconverter)
df_all['Month'] = df_all['New_Date'].apply(monthconverter)
df_all['Day'] = df_all['New_Date'].apply(dayconverter)
```
