```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
# 0.) open selenium
driver = webdriver.Chrome('/usr/local/bin/chromedriver')

# 1.) read csv-file
df = pd.read_csv('schweizerraete.csv')

# 2.) create column Google Hits
df['Google Hits']='0'
```

```python
#driver.close()
```

```python
# 4.) for every entry in list go to google and enter name with selenium
df2 = df.head(5)
for name in df2['Councillors']:
    driver.get("https://www.google.ch/webhp")
    temp = driver.find_element_by_id('lst-ib')
    #print(temp)
    temp.send_keys(name)
    temp.send_keys(Keys.RETURN)
    result_text = driver.find_element_by_id('resultStats')
    result_text = result_text.text[9:-28].replace("’","")
    df2.loc[df2['Councillors']==name, 'Google Hits'] = result_text 
df2
```

```python
driver.close()
```
